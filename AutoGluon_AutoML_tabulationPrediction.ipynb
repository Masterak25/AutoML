# TabularPrediction(Regression) with AutoGluon

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
boston= load_boston()
boston.keys()
print(boston.DESCR)#use this command to know more about the dataset
```

#create a dataframe from the dataset

```python
df=pd.DataFrame(data=boston.data,columns=boston.feature_names)
df.head()
```

# adding price column to the dataframe

```python
df['PRICE'] = boston.target
df.head()
```

# Splitting the dataset

```python
from sklearn.model_selection import train_test_split
df_train,df_test = train_test_split(df, test_size=0.33, random_state=1)
df_train.shape,df_test.shape
```

```python
test_data=df_test.drop(['PRICE'],axis=1)
test_data.head()
```

```python
#! pip install autogluon
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor #to handle tabular data and train models
predictor= TabularPredictor(label ='PRICE').fit(train_data = df_train, verbosity = 2,presets='best_quality')
predictor.leaderboard(df_train, silent=True)
```

# Making Predictions

```python
y_pred = predictor.predict(test_data)
```

```python
y_pred
```

# TabularPrediction(Classification) for healthcare dataset stroke with AutoGluon

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
from sklearn.model_selection import train_test_split #splitting the dataset
from autogluon.tabular import TabularDataset, TabularPredictor #to handle tabular data and train models
df=pd.read_csv('healthcare-dataset-stroke-data.csv') # path to colab notebook #replace this with the path to your dataset
```

# split into train and test sets

```python
df_train,df_test=train_test_split(df,test_size=0.33,random_state=1)
df_train.shape,df_test.shape
```

```python
df.head()
```

```python
test_data=df_test.drop(['stroke'],axis=1)
test_data.head()
```

```python
predictor= TabularPredictor(label ='stroke').fit(train_data = df_train, verbosity = 2,presets='best_quality')
```

```python
predictor.fit_summary()
```

In this case, AutoGluon trained 24 models but we would be more interested to find out which is the best model as selected by AutoGluon. To display this, simply use the leaderboard() command which ranks the trained models in order.

```python
predictor.leaderboard(df_train, silent=True)
```

Additionally, we can also check for the feature importance 

```python
predictor.feature_importance(data=df_train)
```

```python
y_pred = predictor.predict(test_data)
y_pred=pd.DataFrame(y_pred,columns=['stroke'])
y_pred #print the DataFrame
```

evaluation metric accuracy

```python
predictor.evaluate(df_test)
```
